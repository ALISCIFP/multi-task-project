## Joint Plots

```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
pos_train = pd.read_csv('../../data/joint/March20/accuracy/train_pos_stats.txt', header=None, sep=' ').as_matrix()
pos_valid = pd.read_csv('../../data/joint/March20/accuracy/valid_pos_stats.txt', header=None, sep=' ').as_matrix()

plt.plot(1-pos_train,label='train')
plt.plot(1-pos_valid,label='valid')
plt.ylabel('Error')
plt.xlabel('epoch number')
plt.title('POS Accuracy- Joint')
plt.legend()
plt.show()
```
